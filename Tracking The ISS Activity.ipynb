```python
import requests as rq
import matplotlib.pyplot as plt
import time
import json
import numpy as np
ping_counter = 0
y = 0
longitude = []
latitude = []
time_stamp = []
while ping_counter < 100:
    y = 0
    while y == 0:
        x = rq.get('http://api.open-notify.org/iss-now.json').text
        x = json.loads(x)
        latitude.append(float(x['iss_position']['latitude']))
        longitude.append(float(x['iss_position']['longitude']))
        time_stamp.append(x['timestamp'])
        y = 1
        ping_counter += 1
        time.sleep(10)
    else:
        continue
print("longitude: " + str(longitude))
print("latitude: " + str(latitude))
print("timestamps: " + str(time_stamp))
```

```python
plt.plot(latitude,longitude)
plt.xlabel('latitude')
plt.ylabel('longitude')
plt.plot(latitude, longitude, marker="o", markersize=3)
plt.xticks(np.arange(min(latitude), max(latitude), 5))
plt.yticks(np.arange(min(longitude), max(longitude), 20))
plt.show()
```
